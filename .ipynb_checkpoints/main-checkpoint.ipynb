```python
# function to read the images by taking there path

images = list()

path = "/mnt/d/manas-taskphase/mapthemap/"
files = [i for i in os.listdir(path) if os.path.isfile(os.path.join(path,i)) and 'map1snip' in i]
for file in files:
    image = cv2.imread(file)
    images.append(image)


# for i in range(0, len(images)):
#     images[i] = cv2.cvtColor(images[i], cv2.COLOR_BGR2GRAY)

reference_image = images[0]

sift = cv2.SIFT_create()
kp = list()
des = list()
for i in range(0, len(images)):
    # find the keypoints and descriptors with ORB
    thiskp, thisdes = sift.detectAndCompute(images[i], None)
    kp.append(thiskp)
    des.append(thisdes)
    
# BFMatcher with default params
bf = cv2.BFMatcher()
MIN_MATCH_COUNT = 20
```

```python
# BFMatcher with default params
matchedimages = list()
# goodmatches = list()
homographydatas = list()
validtransforms = list()
plt.figure()
# print(kp)
#subplot(r,c) provide the no. of rows and columns
f, axarr = plt.subplots(len(images), len(images))

for i in range(0, len(images)):
    # cv.drawMatchesKnn expects list of lists as matches.
    matchedimages.append(list())
    homographydatas.append(list())
    validtransforms.append(list())
    for j in range(0, len(images)):
        matches = bf.knnMatch(des[i],des[j],k=2)
        # Apply ratio test
        # good = [[0,0] for p in range(len(matches))] 
        # ratio test as per Lowe's paper 
        # for k,(m,n) in enumerate(matches): 
        #     if m.distance < 0.75*n.distance: 
        #         good[k]=[m] 
        good = list()
        # for m,n in matches:
        #     if m.distance < 0.75*n.distance:
        #         good.append([m])
        for match in matches:
            if match[0].distance < 0.75*match[1].distance:
                good.append([match[0]])
        matchedimage = cv2.drawMatchesKnn(images[i],kp[i],images[j],kp[j],good,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
        matchedimages[i].append(matchedimage)
        axarr[i][j].imshow(matchedimages[i][j])
        # for m,n in matches:
        #     if m.distance < 0.75*n.distance:
        #         good.append([m])
        good = []
        for m,n in matches:
            if m.distance < 0.7*n.distance:
                good.append(m)
        if len(good)>MIN_MATCH_COUNT:
            validtransforms[i].append(True)    
            # src_pts = np.float32([kp[i][m.queryIdx].pt for m in good]).reshape(-1,1,2)
            # dst_pts = np.float32([kp[j][m.trainIdx].pt for m in good]).reshape(-1,1,2)
            # homographydata = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
            # homographydatas[i].append(homographydata)
        else:
            print( "{}->{}, Not enough matches are found - {}/{}".format(i,j,len(good), MIN_MATCH_COUNT) )
            validtransforms[i].append(False)
            #homographydatas[i].append(None)



plt.show()

# print(homographydatas)
```

```python
# # homography_matrix[i][j] should contain transform matrix of image[i] to image[j]
# MIN_MATCH_COUNT = 10
# if len(good)>MIN_MATCH_COUNT:
#     src_pts = np.float32([ kp[j][m[0].queryIdx].pt for m in goodmatches]).reshape(-1,1,2)
#     dst_pts = np.float32([ kp[i][m[0].trainIdx].pt for m in goodmatches]).reshape(-1,1,2)
#     M, mask = cv.findHomography(src_pts, dst_pts, cv.RANSAC,5.0)
#     # Finding homography matrix mask
#     matchesMask = mask.ravel().tolist()
#     h,w = images[j].shape
#     pts = np.float32([[0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
#     dst = cv.perspectiveTransform(pts,M) # requires pts and M. store M in matrix 2d list
#     img2 = cv.polylines(img2,[np.int32(dst)],True,255,3, cv.LINE_AA)
# else:
#     print( "Not enough matches are found - {}/{}".format(len(good), MIN_MATCH_COUNT) )
#     matchesMask = None

# draw_params = dict(matchColor = (0,255,0), # draw matches in green color
#  singlePointColor = None,
#  matchesMask = matchesMask, # draw only inliers
#  flags = 2)
# img3 = cv.drawMatches(img1,kp1,img2,kp2,good,None,**draw_params)
# plt.imshow(img3, 'gray'),plt.show()

def getpaddedimg(new_image_height, new_image_width, old_image_width, old_image_height, img, channels = 3):
    color = (0,0,0)
    result = np.full((new_image_height,new_image_width, channels), color, dtype=np.uint8)
    x_center = (new_image_width - old_image_width) // 2
    y_center = (new_image_height - old_image_height) // 2

    # copy img image into center of result image
    result[y_center:y_center+old_image_height, 
        x_center:x_center+old_image_width] = img
    return result
# create new image of desired size and color (blue) for padding



print(validtransforms)
nomatch_indexes = list()


for i in range(0, len(images)):
    thisvalid = False
    for j in range(0, len(images)):
        if i==j:
            continue
        if validtransforms[j][i] is True:
            thisvalid = True
            break
    if thisvalid is False:
        nomatch_indexes.append(i)

ref = False
for i in range(0, len(images)):
    for j in range(0, len(images)):
        if i==j:
            continue
        if validtransforms[j][i] is True:
            reference_image = np.array(images[i], copy = True)
            ref = i
            break
    if ref != False:
        break

unblended_image_indexes = [i for i in range(0,len(images))]
unblended_image_indexes.remove(ref)
for i in nomatch_indexes:
    unblended_image_indexes.remove(i)    

while len(unblended_image_indexes)!=0:
    print(unblended_image_indexes)
    # refkp, refdes = sift.detectAndCompute(reference_image, None)
    for k in unblended_image_indexes:
        new_image_width = reference_image.shape[1] + images[k].shape[1]
        new_image_height = reference_image.shape[0] + images[k].shape[0]
        reference_image = getpaddedimg(new_image_height, new_image_width, reference_image.shape[1], reference_image.shape[0], reference_image)
        current_image = getpaddedimg(new_image_height, new_image_width, images[k].shape[1], images[k].shape[0], images[k])
        
        thiskp, thisdes = sift.detectAndCompute(current_image, None)
        refkp, refdes = sift.detectAndCompute(reference_image, None)
        
        matches = bf.knnMatch(thisdes,refdes,k=2)
        good = list()
        for match in matches:
            if match[0].distance < 0.75*match[1].distance:
                good.append([match[0]])
        # matchedimage = cv2.drawMatchesKnn(images[k],kp[k],reference_image,refkp,good,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
        matchedimage = cv2.drawMatchesKnn(current_image, thiskp, reference_image,refkp,good,None,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

        # axarr[i][j].imshow(matchedimages[i][j])
        # for m,n in matches:
        #     if m.distance < 0.75*n.distance:
        #         good.append([m])
        good = []
        for m,n in matches:
            if m.distance < 0.7*n.distance:
                good.append(m)
        if len(good)>=MIN_MATCH_COUNT:
            src_pts = np.float32([thiskp[m.queryIdx].pt for m in good]).reshape(-1,1,2)
            dst_pts = np.float32([refkp[m.trainIdx].pt for m in good]).reshape(-1,1,2)
            M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
            # matchesMask = mask.ravel().tolist()
            # h,w = images[k].shape
            # pts = np.float32([[0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
            # dst = cv2.perspectiveTransform(pts,M) # requires pts and M. store M in matrix 2d list
            # img2 = cv.polylines(img2,[np.int32(dst)],True,255,3, cv.LINE_AA)
            # Warp the first image using the homography
            # warped = cv2.warpPerspective(images[k], M, (reference_image.shape[1], reference_image.shape[0]))    
            # # Blending the warped image with the second image using alpha blending
            # alpha = 0.5  # blending factor
            # pts = np.float32([[0,0],[0,-1],[-1,-1],[-1,0]]).reshape(-1,1,2)
            # dst = cv2.perspectiveTransform(pts,M)
            # dst = cv2.warpPerspective(img1,H,(img2.shape[1] + img1.shape[1], img2.shape[0])) #wraped image
            # print("dst 1:", dst)
            # dst[0:img2.shape[0], 0:img2.shape[1]] = img2 #stitched image
            # print("dst 2:",dst)
            # reference_image = cv2.addWeighted(dst, alpha, reference_image, 1 - alpha, 0)
            
            # transf = cv2.getPerspectiveTransform(src_pts, dst_pts)
            
            # dst_pad, warped = warpPerspective(src, dst, transf)
            stitched_image = cv2.warpPerspective(current_image, M, (reference_image.shape[1], reference_image.shape[0]))
            alpha = 1.0
            beta = 1.0
            # reference_image = cv2.addWeighted(stitched_image, alpha, reference_image, beta, 0.0)
            reference_image = cv2.bitwise_or(reference_image, stitched_image)
            # cv2.imshow(f"stitch {k}",reference_image)
            # while True:
            #     key = cv2.waitKey(0) & 0xFF
            #     print(k)
            #     if key == ord('c'): # you can put any key here
            #         cv2.destroyAllWindows()
            #         break

            unblended_image_indexes.remove(k) 
        else:
            print( "{}->ref, Not enough matches are found - {}/{}".format(k,len(good), MIN_MATCH_COUNT) )
            homographydata = None
        break

# Removing pepper noise developed during bitwise or
reference_image = cv2.medianBlur(reference_image, 3)

cv2.imshow('stitched', reference_image)
while True:
    key = cv2.waitKey(0) & 0xFF
    print(k)
    if key == ord('c'): # you can put any key here
        cv2.destroyAllWindows()
        break
```
